# GPT-2

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import pickle

def plot__two_perplexity_ratio_distributions__zephyr(n_epochs, upper_lim):
  # Path to the pickle file in Google Drive
  directory_path = '/content/drive/My Drive/zephyr_perplexity_ratios'
  file_path = directory_path + '/perplexity_ratios_full__' + str(n_epochs) +  '_epochs.pkl'

  # Load the lists from the pickle file
  with open(file_path, 'rb') as f:
      stereotyping_demeaning__perplexity_ratios_full, nonstereotyping_nondemeaning__perplexity_ratios_full = pickle.load(f)

  # The variables are now restored
  print(stereotyping_demeaning__perplexity_ratios_full)
  print(nonstereotyping_nondemeaning__perplexity_ratios_full)

  # Define the bin edges, making sure the last bin includes values greater than 100
  step = upper_lim/100
  bins = np.arange(0, upper_lim+step+0.001, step)
  print('bins:', bins)

  #Turn all values larger than 1000 to 1000.1 (so that the last bin contain all those values)
  stereotyping_demeaning__perplexity_ratios_full__for_plot = [(upper_lim+0.01) if x > upper_lim else x for x in stereotyping_demeaning__perplexity_ratios_full]
  nonstereotyping_nondemeaning__perplexity_ratios_full__for_plot = [(upper_lim+0.01) if x > upper_lim else x for x in nonstereotyping_nondemeaning__perplexity_ratios_full]

  plt.figure(figsize=(12, 6))

  # Plot the histogram for stereotyping_demeaning_perplexity_ratios
  sns.histplot(stereotyping_demeaning__perplexity_ratios_full__for_plot, bins=bins, kde=True, color='red', label='Harmful (Stereotyping and Demeaning)', stat="probability", alpha=0.5)

  # Plot the histogram for nonstereotyping_nondemeaning_perplexity_ratios
  sns.histplot(nonstereotyping_nondemeaning__perplexity_ratios_full__for_plot, bins=bins, kde=True, color='green', label='Harmless (Non-stereotyping and Non-demeaning)', stat="probability", alpha=0.5)

  # Add titles and labels
  plt.title('Histogram of Zephyr Perplexity Ratios on Harmful and Harmless Subsets of FairPrism (' + str(n_epochs) + ' epochs)')
  plt.xlabel('Perplexity Ratios. Note: the last column shows all perplexity ratios that are above the plotting limit (> ' + str(upper_lim) + ')')
  plt.ylabel('Probability Density')


  # Add a legend
  plt.legend()


  # Define the directory path
  directory_path = '/content/drive/My Drive/zephyr_perplexity_ratios'
  plt.savefig(directory_path  + '/zephyr__perplexity_ratio_distributions_' + str(n_epochs) +  '_epochs.pdf')

  # Show the plot
  plt.show()
```

```python
plot__two_perplexity_ratio_distributions__zephyr(n_epochs=4, upper_lim=8)
```

```python
plot__two_perplexity_ratio_distributions__zephyr(n_epochs=8, upper_lim=300)
```
